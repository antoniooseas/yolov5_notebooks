```python
import os
import shutil
```

```python
def createFolder(folder):
    if not os.path.exists(folder):
        os.mkdir(folder)
    else:
        print('pasta existente!')
```

<h2> preparar pastas com os videos para organizacao dos dados

```python
#informar as pastas que estão os videos (observacao: imagens ja no formato da Yolo from CVAT)
root_path = '/Users/Oseas/Desktop/dataset_nestle/'
folders = []
prefixos = []
for folder in os.listdir(root_path):
    full_path_folder = os.path.join(root_path, folder)
    if os.path.isdir(full_path_folder):
        folders.append(full_path_folder+'/obj_train_data/')
        prefixos.append(folder)
```

```python
#folder onde ficarão as imagens e os labels
path_images = f'{root_path}images/'
path_labels = f'{root_path}labels/'

#criar as pastas para images e labels
createFolder(path_images)
createFolder(path_labels)

#contadores para que as imagens tenha numeros crescentes, iniciando de forma automática para nao sobrepor as imagens
cont_images = len([f for f in os.listdir(path_images) if f.endswith('.jpg')])
cont_txt = len([f for f in os.listdir(path_labels) if f.endswith('.txt')])

#fazendo a iteracao na lista de folders
for i, folder in enumerate(folders):
    files = os.listdir(folder)
    for file in files:
        if ".jpg" in file:
            #os.rename(folder+file,f"{path_images}{cont_images}_{prefixos[i]}.jpg")
            shutil.copyfile(folder+file,f"{path_images}{cont_images}_{prefixos[i]}.jpg")
            #print(folder+file)
            #print(f"{path_images}{cont_images}_{prefixos[i]}.jpg")
            cont_images += 1
        elif ".txt" in file:
            #os.rename(folder+file,f"{path_labels}{cont_txt}_{prefixos[i]}.txt")
            shutil.copyfile(folder+file,f"{path_labels}{cont_txt}_{prefixos[i]}.txt")
            #print(folder+file)
            #print(f"{path_labels}{cont_txt}_{prefixos[i]}.txt")
            cont_txt += 1
```

```python
#acessar diretório que contem os txts com as marcacoes
files = os.listdir(path_labels)
for file in files:
    if ".txt" in file:
        with open(os.path.join(path_labels,file), 'r') as arquivo:
            linhas = arquivo.readlines()
        # Substitui o primeiro caractere de cada linha que for igual a 4 por 2
        for i in range(len(linhas)):
            if linhas[i].startswith('2'):
                linhas[i] = '2' + linhas[i][1:]
        with open(os.path.join(path_labels,file), 'w') as arquivo:
            arquivo.writelines(linhas)   
        print('Arquivo alterado -> ' + arquivo.name)
```
